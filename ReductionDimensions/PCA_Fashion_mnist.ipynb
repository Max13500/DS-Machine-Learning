### Récupération du dataset

```python
import kagglehub as kh
import pandas as pd

# Télécharger le dataset dans un dossier local
dataset_ref = "zalando-research/fashionmnist"
path = kh.dataset_download(dataset_ref)

# Localiser les fichiers CSV dans le dossier téléchargé
train_csv_file = f"{path}/fashion-mnist_train.csv"
test_csv_file = f"{path}/fashion-mnist_test.csv"

# Charger le fichier CSV dans un DataFrame
df_train = pd.read_csv(train_csv_file)
df_test = pd.read_csv(test_csv_file)

```

```python
display(df_train)
df_train.info()
```

```python
display(df_test)
df_test.info()
```

```python
# Séparation features / target
X_train = df_train.drop(columns='label')
y_train= df_train['label']
X_test = df_test.drop(columns='label')
y_test= df_test['label']
```

```python
# Affichage d'une image de train et de test
import matplotlib.pyplot as plt

plt.figure(figsize=(1,1))
img = X_train.iloc[0].values.reshape(28,28)
plt.imshow(img,cmap='gray')
plt.title(f"Classe : {y_train[0]}")
plt.show()

plt.figure(figsize=(1,1))
img = X_test.iloc[0].values.reshape(28,28)
plt.imshow(img,cmap='gray')
plt.title(f"Classe : {y_test[0]}")
plt.show()
```

### PCA

```python
from sklearn.decomposition import PCA

# Décomposition PCA sur l'ensemble d'entraînement
pca = PCA()
pca = pca.fit(X_train)

import numpy as np
plt.figure(figsize=(15,7))
plt.subplot(121)
plt.plot(np.arange(len(pca.explained_variance_ratio_))+1,pca.explained_variance_ratio_)
plt.xlabel("Nb de composantes")
plt.ylabel("Part de variance expliquée")

plt.subplot(122)
plt.plot(np.arange(len(pca.explained_variance_ratio_))+1,pca.explained_variance_ratio_.cumsum())
plt.axhline(y = 0.9, color ='r', linestyle = '--')
plt.xlabel("Nb de composantes")
plt.ylabel("Somme cumulative de la variance expliquée")
plt.grid(True)
plt.show()
```

### Nb de composantes

```python
# Décomposition PCA sur 10 axes
pca =PCA(10)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
print ("Part de variance expliquée :", pca.explained_variance_ratio_.sum())

# Entraînement et évaluaton RandomForest avec PCA
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_pca,y_train)
rfc.score(X_test_pca,y_test)
```

```python
# Décomposition PCA avec part de varriance expliquée = 0.9
pca =PCA(0.9)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
print ("Part de variance expliquée :", pca.explained_variance_ratio_.sum())

# Entraînement et évaluaton RandomForest avec PCA
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_pca,y_train)
rfc.score(X_test_pca,y_test)
```

```python

print ("Nb de composantes retenues :", pca.n_components_)

# Affichage des 2 composantes principales
plt.scatter(X_train_pca[:,0],X_train_pca[:,1],c=y_train,cmap=plt.cm.Spectral)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title("Données projetées sur les 2 1ers axes de PCA")
plt.show()
```

### Comparaison avec LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

# Décomposition LDA
lda = LDA()
X_train_lda = lda.fit_transform(X_train,y_train)
X_test_lda = lda.transform(X_test)
print ("Part de variance expliquée :", lda.explained_variance_ratio_.sum())

# Entraînement et évaluaton RandomForest avec LDA
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train_lda,y_train)
rfc.score(X_test_lda,y_test)
```

```python
print ("Nb de composantes retenues :", X_train_lda.shape[1])

# Affichage des 2 composantes principales
plt.scatter(X_train_lda[:,0],X_train_lda[:,1],c=y_train,cmap=plt.cm.Spectral)
plt.xlabel('LDA 1')
plt.ylabel('LDA 2')
plt.title("Données projetées sur les 2 1ers axes de PCA")
plt.show()
```

### Interprétation de la PCA

```python
# Part de la variance expliquée de chaque axe
plt.bar([f"Axe {i+1}" for i in range(10)],pca.explained_variance_ratio_[:10])
plt.xlabel("Axe de la PCA")
plt.ylabel('Part de Variance Expliquée')
plt.show()
```

```python
# Corrélations des 10 1ers axes
df_charges_factorielles = pd.DataFrame(pca.components_[:10],index=[f"Axe {i+1}" for i in range(10)],columns=X_train.columns)
display (df_charges_factorielles)
```

```python
# Cercle de corrélation des 2 premiers axes
def draw_correlation_circle(df_charges_factorielles, arrow_length=0.1, label_rotation=0):
    fig, ax = plt.subplots(figsize=(8, 8))
    for i, var in enumerate(df_charges_factorielles.columns):
        x = df_charges_factorielles.loc['Axe 1', var]
        y = df_charges_factorielles.loc['Axe 2', var]
        ax.arrow(0, 0, x, y, head_width=arrow_length, head_length=arrow_length, fc='gray', ec='gray')
        #ax.text(x*1.15, y*1.15, var, ha='center', va='center', rotation=label_rotation, fontsize=9)
    circle = plt.Circle((0, 0), 1, facecolor='none', edgecolor='black')
    ax.add_artist(circle)
    ax.set_xlim(-1.1, 1.1)
    ax.set_ylim(-1.1, 1.1)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('Axe 1 (PC1)')
    ax.set_ylabel('Axe 2 (PC2)')
    ax.set_title('Cercle des Corrélations')
    plt.grid()
    plt.show()

# Appelez la fonction pour tracer le cercle de corrélation
draw_correlation_circle(df_charges_factorielles)
```
